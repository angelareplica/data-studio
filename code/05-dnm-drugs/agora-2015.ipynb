```python
df = pd.read_csv('Agora.csv', encoding = "utf-8")
```

```python
df.dtypes
```

```python
df['Category'] = df[' Category']
df['Vendor'] = df[' Item']
df['ItemDescription'] = df[' Item Description']
df['Price'] = df[' Price']
df['Origin'] = df[' Origin']
df['Destination'] = df[' Destination']
df['Rating'] = df[' Rating']
df['Remarks'] = df[' Remarks']
```

```python
df.Category.value_counts()
```

```python
drugs_df = df[df.Category.str.contains('Drugs')]
```

```python
drugs_df.Category.value_counts()
```

```python
drugs_df.Category.value_counts().plot(kind='pie', figsize=(10, 10))
centercircle = plt.Circle((0,0),0.5, fc='white',linewidth=1.25)
fig = plt.gcf()
fig.gca().add_artist(centercircle)
plt.show()
```

```python
psychedelics_df = drugs_df[drugs_df.Category.str.contains('Psychedelics')]
psychedelics_df.Category.value_counts().plot(kind='pie', figsize=(10, 10))
centercircle = plt.Circle((0,0),0.5, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centercircle)
plt.show()
```
